```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("climate_change.csv")
dfCanada = pd.read_csv("climate_change_canada.csv")
```

```python
def retrieve_data_from_month(df, month):
    needed_data = df.loc[df[' Statistics'] == (" " + month + " Average")]
    y = needed_data["Temperature - (Celsius)"]
    
    return y
```

```python
def plot_all_month_data(df):
    months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
    for month in months:
        print(month)
        x = np.array(range(1901,2017)).flatten()
        y = retrieve_data_from_month(df, month)
        y = np.array(y).flatten()
        m, b = np.polyfit(x, y, 1)
        plt.xlim(1900,2016)
        plt.plot(x, y, 1)  
        plt.plot(x, m*x + b, 1)
        plt.title("Temperature vs Time in Canada")
        plt.ylabel("Temperature (C)")
        plt.xlabel("Time (Year)")
        print(m)
    plt.savefig("allmonthsCanada.png")

    
```

```python
plot_all_month_data(df)
```

```python
def plot_month_range_data(df, months, season):
    for month in months:
        print(month)
        x = np.array(range(1901,2017)).flatten()
        y = retrieve_data_from_month(df, month)
        y = np.array(y).flatten()
        m, b = np.polyfit(x, y, 1)
        plt.xlim(1900,2016)
        plt.plot(x, y, 1)  
        plt.plot(x, m*x + b, 1) 
        plt.title("Temperature vs Time in Canada in " + season)
        plt.ylabel("Temperature (C)")
        plt.xlabel("Time (Year)")
        print(m)
    plt.savefig(season + "Canada.png")

    
```

```python
plot_month_range_data(df, ["Mar", "Apr", "May"], "Spring")
```

```python
plot_month_range_data(df, ["Jun", "Jul", "Aug"], "Summer")
```

```python
plot_month_range_data(df, ["Sep", "Oct", "Nov"], "Fall")
```

```python
plot_month_range_data(df, ["Dec", "Jan", "Feb"], "Winter")
```
